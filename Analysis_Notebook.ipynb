```python
# First import python packages for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date 
```

#### Example Code: Methane Data through Time
Below is code to import a dataset from a csv file and then plot the data in a graph. Read through the below code to get an understanding of how to do basic data analysis in Python. Then use the code to create your own code for today's lab.

```python
# Import Methane Dataset into table (dataframe), drop any columns with NaN data, say which column has date data
file_name = 'Datasets/Methane_Mauna_Loa_data.csv' # string with file path
CH4 = pd.read_csv(file_name, skiprows=45, parse_dates=['date']).dropna()
```

```python
# View top of the dataframe of Methane Data to check that it looks okay
CH4.head()
```

```python
# Plot the average methane data through time
plt.plot(CH4['date'], CH4['average CH4 (ppb)'],'r') # plt.plot(x data, y data,line color)
plt.xlabel('Date')
plt.ylabel('Methane Concentration (ppb)')
plt.show()
```
