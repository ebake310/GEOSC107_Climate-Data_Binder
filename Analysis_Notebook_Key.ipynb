#### Part IV: Temperature Change & Mass Extinctions
Plot data on the temperature changes that have occured over the last 450 Million years and how this relates to the extinction of different animal species (groups within their families). This part uses two datasets that you must load. Your final figure will plot all three datasets on overlapping axes.

```python
# Import the data using Pandas (pd) function
# Hint: just use pd.read_csv(filename)

# There are two dataset files for this part
file_name4 = 'Datasets/Temperature_extinction_data.csv' # temperature change estimates and rate data
file_name5 = 'Datasets/Extinction_data.csv'# extinction rate data

# Import both datasets into separate dataframes
temp = pd.read_csv(file_name4)
extinc = pd.read_csv(file_name5)
```

```python
# Display top of temperature change and rate dataset
temp.head()
```

```python
# Display top of extinction rate dataset
extinc.head()
```

```python
# Set figure dimensions and axis parameters (done for you)
fig, ax = plt.subplots(figsize=(11, 4))
fig.subplots_adjust(right=0.75)

twin1 = ax.twinx()
twin2 = ax.twinx()

# Offset the right spine of twin2.  The ticks and label have already been placed on the right by twinx above.
twin2.spines.right.set_position(("axes", 1.1))

# Plot the three datasets on three overlapping axes
p1, = ax.plot(temp['Age (Million years ago)'], temp['Temperature Change (Celsius)'], "b--", label="Temperature Change (C)")
p2, = twin1.plot(temp['Age (Million years ago)'], temp['Rate Temp Change (Celsius per million years)'], "r--", label="Rate Temp Change (C/Myr)")
p3, = twin2.plot(extinc['Age (Million years ago)'], extinc['Extinction rate (families per million years)'], "k-", label="Extinction Rate (Families/Myr)")

# Set axis limits
ax.set_xlim(0, 452)
ax.set_ylim(0, 10)
twin1.set_ylim(0, 120)
twin2.set_ylim(0, 20)

# Set axis labels
ax.set_xlabel("Time (Myr)")
ax.set_ylabel("Temperature Change (C)")
twin1.set_ylabel("Rate Temp Change (C/Myr)")
twin2.set_ylabel("Extinction Rate (Families/Myr)")

# Set axis properties
ax.yaxis.label.set_color(p1.get_color())
twin1.yaxis.label.set_color(p2.get_color())
twin2.yaxis.label.set_color(p3.get_color())

tkw = dict(size=4, width=1.5)
ax.tick_params(axis='y', colors=p1.get_color(), **tkw)
twin1.tick_params(axis='y', colors=p2.get_color(), **tkw)
twin2.tick_params(axis='y', colors=p3.get_color(), **tkw)
ax.tick_params(axis='x', **tkw)

# Add legend
ax.legend(handles=[p1, p2, p3])

plt.show()
```
